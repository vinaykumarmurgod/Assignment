1.Discuss string slicing and provide example.

Ans: String slicing is a method in Python to extract a part of a string by specifying a range of indices. The syntax for slicing is string[start:end:step].

* start: The position where the slice begins (inclusive).
* end: The position where the slice ends (exclusive).
* step: (Optional) The interval between elements.

 Example: If we have the string "Hello, World!":

* string[0:5] will return "Hello" (characters from index 0 to 4).
* string[7:] will return "World!" (characters from index 7 to the end).

Slicing is useful when you need to extract parts of strings for operations like substring searches or formatting.


2.Explain the key features of lists in Python.

Ans: A list in Python is an ordered, mutable collection of elements. Key features include:

* Ordered: Items in a list are stored in a specific sequence, and you can access them using their index.
* Mutable: You can modify the list by adding, changing, or removing elements.
* Allows duplicates: Lists can contain the same item multiple times.
* Supports different data types: A list can hold elements of different types, such as integers, strings, and other lists.


Example: A list can look like this: my_list = [1, "apple", 3.14, True]. You can access the first element with my_list[0] which returns 1.



3.Describe how to access, modify, and delete elements in a list with examples

Ans:
* Access Elements: Use the index to retrieve elements from a list. Lists are zero-indexed.
 * Example: my_list = [10, 20, 30], my_list[1] returns 20.
* Modify Elements: Assign a new value to a specific index.
 * Example: my_list[1] = 25 changes the second element to 25.
*Delete Elements: You can delete elements using del, pop(), or remove().
 * Example: del my_list[1] removes the element at index 1. Now my_list = [10, 30].


4.Compare and contrast tuples and lists with examples.

Ans:
* Tuples:

 * Definition: Tuples are immutable sequences, meaning their elements cannot be changed after creation.
 * Example: A tuple like (3, 5) represents fixed coordinates.
* Lists:

 * Definition: Lists are mutable sequences, allowing changes to their elements after creation.
 * Example: A shopping list like [“apples,” “bananas,” “oranges”] can be modified as needed.
* Key Differences:

 * Syntax: Tuples use parentheses (); lists use square brackets [].
 * Mutability: Tuples are immutable; lists are mutable.

5.Describe the key features of sets and provide examples of their use.

Ans:
* Definition: Sets are unordered collections of unique elements that automatically eliminate duplicates.

 * Example: A set like {1, 2, 3, 4} will store only unique values, ignoring any duplicates.
* Unordered: The elements in a set do not have a specific order, meaning they cannot be accessed by index.

 * Example: In the set {3, 1, 2}, the order may change each time you print it.
* Mutable: Sets are mutable, allowing you to add or remove elements after creation.

 * Example: Starting with {“apple,” “banana”}, you can add “orange” to create {“apple,” “banana,” “orange”}.
* No Duplicates: Sets automatically remove duplicate entries, ensuring each element is unique.

 * Example: If you try to create a set with {1, 2, 2, 3}, it will store it as {1, 2, 3}.

6.Discuss the use cases of tuples and sets in Python programming.

Ans:
* Tuples: Tuples are often used to store data that should not change, such as coordinates or configuration data. Their immutability ensures that the values stay the same throughout the program.
 * Example: A tuple (x, y) can represent a 2D coordinate.
* Sets: Sets are useful when you need to work with unique elements or perform mathematical set operations like union, intersection, or difference.
 * Example: Use sets to remove duplicates from a list or find common elements between two collections, like set1 & set2 for the intersection.

7.Describe how to add, modify, and delete items in a dictionary with examples.

Ans:A dictionary is a collection of key-value pairs where each key must be unique.

* Add Items: You can add a new key-value pair by assigning a value to a new key.
 * Example: my_dict['name'] = 'Alice' adds the key 'name' with the value 'Alice'.
* Modify Items: To modify a value, assign a new value to an existing key.
 * Example: my_dict['age'] = 30 changes the value associated with the key 'age'.
* Delete Items: Use del or pop() to remove a key-value pair.
 * Example: del my_dict['name'] removes the key 'name' from the dictionary.

8.Discuss the importance of dictionary keys being immutable and provide examples.

Ans: Dictionary keys must be immutable because Python uses hashing to manage dictionaries internally. Hash values must remain constant, which is why mutable types like lists cannot be used as keys.

* Immutable keys: Strings, numbers, and tuples are commonly used as keys because their values do not change.
 * Example: You can use a string as a key: my_dict = {"name": "Alice"}.
* Mutable keys not allowed: Lists and other mutable types are not allowed because they can change after being used as a key, causing issues with the dictionary's hash-based structure.
