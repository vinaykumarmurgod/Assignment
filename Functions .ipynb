**Theory Questions**


 1. What is the difference between a function and a method in Python?

* A function is a standalone block of code designed to perform a specific task. It can be called independently and does not belong to any object.
* A method is a function that is associated with an object and is called on that object. For example, if you have a string object, you can call its method to manipulate the string.
* Example: len() is a function that returns the length of an object, while "Hello".upper() is a method that converts the string to uppercase.



 2. Explain the concept of function arguments and parameters in Python.

* Parameters are the variables defined in a function's definition. They act as placeholders for the values that will be passed to the function.
* Arguments are the actual values you provide to the function when you call it.
* Example: In a function def greet(name), name is a parameter. When you call greet("Alice"), "Alice" is the argument.

 3. What are the different ways to define and call a function in Python?

* You can define a function using the def keyword followed by the function name and parameters.
* Functions can be called using their name followed by parentheses and any necessary arguments.
* Example: A function defined as def add(a, b) can be called using add(2, 3).

 4. What is the purpose of the `return` statement in a Python function?

* The return statement is used to exit a function and send back a value to the caller. It allows a function to produce output that can be used later.
* Example: A function that calculates the square of a number might return the squared value to the caller.

 5. What are iterators in Python and how do they differ from iterables?

* An iterable is an object that can be looped over, like lists and strings. It implements the __iter__() method.
* An iterator is an object that keeps track of its current position during iteration. It implements the __next__() method to access the next item.
* Example: A list is iterable; you can use a loop to go through its elements. The list itself can produce an iterator that retrieves elements one at a time.



 6. Explain the concept of generators in Python and how they are defined.

* Generators are a special type of iterator created using a function that contains yield statements. They allow you to iterate through a sequence of values without storing them all in memory.
* Example: A generator can produce a sequence of numbers one at a time, pausing each time it yields a value until the next value is requested.

7. What are the advantages of using generators over regular functions?

* Generators are more memory-efficient than regular functions because they yield one item at a time instead of returning a complete list. This is especially useful for processing large datasets.
* They provide a simple way to create iterators with less code.
* Example: A generator for Fibonacci numbers can generate each number in the sequence on-the-fly, without generating the entire sequence upfront.

 8. What is a lambda function in Python and when is it typically used?

* A lambda function is a small anonymous function defined using the lambda keyword. It can take any number of arguments but can only have one expression.
* Example: You might use a lambda function to define a quick operation, like squaring a number, in a concise way.

 9. Explain the purpose and usage of the `map()` function in Python.

* The map() function applies a specified function to each item in an iterable (like a list) and returns a map object (an iterator). It is often used for transforming data.
* Example: You could use map() to convert a list of strings into a list of their lengths.

 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

* map() applies a function to every item in an iterable and returns a new iterable with the results.
* reduce() takes a function and an iterable, applying the function cumulatively to the items to reduce them to a single value.
* filter() creates a new iterable containing only the items that satisfy a specified condition.
* Example:

 * map() could be used to square each number in a list.
 *reduce() could sum up all numbers in a list to get a total.
 * filter() could be used to keep only even numbers from a list.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def even(a):
  esum = 0
  for i in a:
    if i%2==0:
      esum += i
  return esum
```

```python
even([3,47,101,24,56,25,0,-7,-46])
```

2. Create a Python function that accepts a string and returns the reverse of that string.

```python
def reverse(b):
  reversed_b = b[::-1]
  return reversed_b
```

```python
reverse("Hello")
```

 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
def square(c):
  l = []
  for i in c:
    l.append(i*i)
  return l
```

```python
square([3,45,65,34,-8])
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 1 or n > 200:
        a = print("Please enter a number between 1 and 200")
        return a
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
is_prime(73)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
```

```python
for num in fibonacci(10):
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
for power in powers_of_two(5):
    print(power)
```

 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    file = open(file_path, 'r')
    for line in file:
        yield line.strip()
    file.close()
```

```python
for line in read_file_line_by_line('questions.txt'):
    print(line)
```

 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
tuples_list = [(1, 'banana'), (3, 'apple'), (2, 'cherry'), (4, 'date')]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)
```

 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
z = "apple"
vowels = "aeiouAEIOU"
filtered_chars = filter(lambda x: x not in vowels, z)
filtered_list = list(filtered_chars)
print("Filtered characters:", filtered_list)
print("Filtered string:", ''.join(filtered_list))
```
